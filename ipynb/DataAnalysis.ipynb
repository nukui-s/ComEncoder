```python
import csv
import pandas as pd
```

```python
with open("data/com-dblp.top5000.cmty.txt", "r") as f:
    reader = csv.reader(f, delimiter="\t")
    coms = {}
    for i, line in enumerate(reader):
        coms[i] = [int(c) for c in line]
```

```python
len(coms)
```

```python
n_coms = {i: len(com) for i, com in coms.items()}
```

```python
nc = list(reversed(sorted(n_coms.values())))
```

```python
target_nodes = []
for i, com in coms.items():
    if len(com) < 100: continue
    target_nodes += com
```

```python
len(target_nodes)
```

```python
target_nodes = frozenset(target_nodes)
```

```python
target_nodes
```

```python
with open("data/com-dblp.ungraph.txt") as f:
    reader = csv.reader(f, delimiter="\t")
    edge_list = []
    for s, d in reader:
        s = int(s)
        d = int(d)
        if s in target_nodes and d in target_nodes:
            edge_list.append((s, d))
```

```python
len(edge_list)
```

```python
target_nodes
```

```python
encoded_nodes = {n:i for i, n in enumerate(target_nodes)}
```

```python
encoded_nodes
```

```python
with open("data/dblp_cleaned_com.txt", "w") as f:
    writer = csv.writer(f, delimiter="\t")
    cnt = 0
    for i, com in coms.items():
        if len(com) < 100: continue
        for c in com:
            writer.writerow([encoded_nodes[c], cnt])
        cnt += 1
```

```python
with open("data/dblp_cleaned_edges.txt", "w") as f:
    writer = csv.writer(f, delimiter="\t")
    for s, d in edge_list:
        writer.writerow([encoded_nodes[s], encoded_nodes[d]])
```

```python
n_cnt = {}
for s, d in edge_list:
    n_cnt[s] = n_cnt.get(s, 0) + 1
    n_cnt[d] = n_cnt.get(d, 0) + 1
```
